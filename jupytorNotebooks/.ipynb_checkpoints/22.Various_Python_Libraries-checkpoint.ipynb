## time

- time.time()  
    - UTC(Universal Time Coordinated 협정 세계 표준시)를 사용하여 현재 시간을 실수 형태로 돌려주는 함수  
    - 1970년 1월 1일 0시 0분 0초를 기준으로 지난 시간을 초 단위로 돌려준다
    
    
- time.localtime  
    - time.time()이 돌려준 실수 값을 사용해서 연도, 월, 일, 시, 분, 초, ... 의 형태로 바꾸어 주는 함수
    
    
    
- time.strftime('출력할 형식 포맷 코드', time.localtime(time.time()))

```python
import time

time.time()
```

```python
time.localtime(time.time())
```

```python
time.strftime('%x', time.localtime(time.time()))
```

```python
time.strftime('%c', time.localtime(time.time()))
```

## image 를 읽어서 display 

- PIL (Python Image Library)
    - Image.open() : image file open
    

- internet 상의 image 는 byte code 로 전송되므로 BytesIO 를 이용하여 binary 로 변환  

```python
import requests
from PIL import Image
from io import BytesIO

def load_image(img_path):
    
    if "http" in img_path:
        response = requests.get(img_path)
        image = Image.open(BytesIO(response.content))
    else:
        image = Image.open(img_path)
    
    return image
```

```python
url = "https://hips.hearstapps.com/hmg-prod.s3.amazonaws.com/images/golden-retriever-royalty-free-image-506756303-1560962726.jpg?crop=0.672xw:1.00xh;0.166xw,0&resize=640:*"
```

```python
load_image(url)
```

```python
img = load_image("images/dog.0.jpg")
img
```

```python
print("Format: {0}\nSize: {1}\nMode: {2}".format(img.format, img.size, img.mode))
```

### PIL 에 사전 정의된 Image Filter 를 이용하여 image 변환 
- ImageFilter.BLUR   
- CONTOUR  
- DETAIL  
- EDGE_ENHANCE  
- EDGE_ENHANCE_MORE  
- EMBOSS  
- FIND_EDGES  
- SHARPEN  
- SMOOTH  
- SMOOTH_MORE  

```python
from PIL import ImageFilter

img.filter(ImageFilter.BLUR)
```

```python
img.filter(ImageFilter.CONTOUR)
```

```python
img.filter(ImageFilter.EMBOSS)
```

### PIL 을 이용한 image 자르기

img.crop((left, upper, right, lower))  - left/upper corner, right/lower corner  

ex) (800, 600) pixel --> (0, 0), (800, 600)

```python
cropped = img.crop((100, 50, 400, 300))    # cropping (left, upper, right, lower) pixel coordinates
cropped
```

### image save  

```python
cropped.save('images/cropped-png.png')
```

```python
img = Image.open('images/cropped-png.png')
print(img.format)
```
